11. Container With Most Water

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

```python
class Solution(object):
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        l = 0
        r = len(height) - 1
        max_volume = 0

        while l < r:
            act_volume = (r-l) * min(height[l], height[r])
            max_volume = max(act_volume, max_volume)           
                
            if height[l] <= height[r]:
                l+=1
            else:
                r-=1

        return max_volume
```

```python
def solution(height):

    def calc_volume(idx_l, idx_r):
        w = idx_r - idx_l
        h = min(height[idx_l], height[idx_r])
        return w * h

    max_volume = 0
    for l in range(len(height)-1):
        for r in range(l, len(height)):
            act_volume = calc_volume(l, r)
            if act_volume > max_volume:
                max_volume = act_volume

    return max_volume
    
```
