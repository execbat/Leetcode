94. Binary Tree Inorder Traversal

Given the root of a binary tree, return the inorder traversal of its nodes' values.

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def inorderTraversal(self, root):
        if root is None:
            return []
        return self.inorderTraversal(root.left) + [root.val] + self.inorderTraversal(root.right)
        
```

```python
input = [1,2,3,4,5,"null",8,"null","null",6,7,9]
```

```python
# Collection the Tree from list description
from collections import deque

def build_tree_by_levels(list_obj):
    if len(list_obj) == 0:
        return []

    
    q = deque()

    root = TreeNode(list_obj[0])
    q.append(root)
    i = 0
    while q and i < len(list_obj):
        node = q.popleft()
        i += 1

        if i < len(list_obj):
            if list_obj[i] != 'null':
                node.left = TreeNode(list_obj[i])
                q.append(node.left)

        i += 1 
        if i < len(list_obj):
            if list_obj[i] != 'null':
                node.right = TreeNode(list_obj[i])
                q.append(node.right)

    return root

tree = build_tree_by_levels(input)    
tree
  
```

```python
def inorderTraversal(tree_node):
    if tree_node is None:
        return []
    return inorderTraversal(tree_node.left) + [tree_node.val] + inorderTraversal(tree_node.right)

inorderTraversal(tree)
```
