2. Add Two Numbers

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def addTwoNumbers(self, l1, l2, transfer = 0):
        def get_val_next(linked_l):
            if linked_l:
                return linked_l.val, linked_l.next            
            else:
                return 0, None

        if not l1 and not l2:
            if transfer != 0:
                return ListNode(transfer)
            return None
        
        val_1, next_1 = get_val_next(l1)
        val_2, next_2 = get_val_next(l2)

        sum = val_1 + val_2 + transfer
        new_val = sum % 10
        transfer = sum // 10

        res_node = ListNode(new_val)
        res_node.next = self.addTwoNumbers(next_1, next_2, transfer)

        return res_node
```

```python
l1 = [9,9,9,9,9,9,9]
l2 = [9,9,9,9]

# Output: [8,9,9,9,0,0,0,1]
```

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# help function to build a linked list
def build_linked_list(l):
    if not l:
        return None
    
    linked_list = ListNode(l[0])
    linked_list.next = build_linked_list(l[1:])
    return linked_list

# help function to make list representation of linked list
def list_represent(linked_l):
    if not linked_l:
        return []

    res = [linked_l.val]
    node = linked_l
    while True:
        if  node.next:
            res.append(node.next.val)
            node = node.next
        else:
            break

    return res
    
```

```python
linked_l1 = build_linked_list(l1)
linked_l2 = build_linked_list(l2)
```

```python
print(list_represent(linked_l1))
print(list_represent(linked_l2))
```
