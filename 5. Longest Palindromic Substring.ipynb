5. Longest Palindromic Substring

Given a string s, return the longest 
palindromi  
substr g  in s.

```python
s = "babad"
```

```python
def longest_palindrome(s):
    def expand_from_center(left, right):
        while left >=0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1

        return s[left + 1 : right ] # left + 1 because cycle breaks on inequation
        
    longest = ""
    for i in range(len(s)):
        even_res = expand_from_center(i,i+1)
        odd_res = expand_from_center(i, i)

        if len(even_res) > len(longest):
            longest = even_res
        if len(odd_res) > len(longest):
            longest = odd_res  
            
    return longest
    

longest_palindrome(s)    
```
