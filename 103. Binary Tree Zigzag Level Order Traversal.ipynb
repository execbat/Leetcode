103. Binary Tree Zigzag Level Order Traversal

Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

```python
class Solution(object):
    def zigzagLevelOrder(self, tree):
        if not tree:
            return []

        zigzag_traversal = []
        q = deque()
        q.append(tree)
        forward = 1
        while q:
            level = []
            obj_in_level = len(q)
            
            for _ in range(obj_in_level):
                node = q.popleft()
                level.append(node.val)
                if node.left:
                    q.append(node.left)

                if node.right:
                    q.append(node.right)

            
            zigzag_traversal.append(level[::forward])
            forward *= -1

        return zigzag_traversal
```

```python
# Input 
root = [3,9,20,"null","null",15,7]
```

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# Help function to build a tree from list representation

from collections import deque

def build_tree_by_levels(list_obj):
    if len(list_obj) == 0:
        return []

    tree = TreeNode(list_obj[0])
    q = deque()
    q.append(tree)
    i = 1
    while q and i < len(list_obj):
        node = q.popleft()
        if list_obj[i] != "null":
            node.left = TreeNode(list_obj[i])
            q.append(node.left)

        i+=1
        if i < len(list_obj):
            if list_obj[i]!= "null":
                node.right = TreeNode(list_obj[i])
                q.append(node.right)

        i+=1

    return tree

    
tree = build_tree_by_levels(root)
tree
```

```python
def zigzag_level_traversal(tree):
    if not tree:
        return []

    zigzag_traversal = []
    q = deque()
    q.append(tree)
    forward = 1
    while q:
        level = []
        obj_in_level = len(q)
        
        for _ in range(obj_in_level):
            node = q.popleft()
            level.append(node.val)
            if node.left:
                q.append(node.left)

            if node.right:
                q.append(node.right)

        
        zigzag_traversal.append(level[::forward])
        forward *= -1

    return zigzag_traversal
```
