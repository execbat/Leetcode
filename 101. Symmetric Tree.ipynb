101. Symmetric Tree

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

```python
class Solution(object):
    def isSymmetric(self, root):
        def isMirror(l, r):
            if not l and not r: # both thees are None
                return True

            if not l or not r: # one of them is None
                return False 

            return l.val == r.val and isMirror(l.right, r.left) and isMirror(l.left, r.right)

        return isMirror(root.left, root.right)
```

```python
# root = [1,2,2,3,4,4,3]
root = [1,2,2,"null",3,"null",3]
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# Help function to build tree from list representation by levels
from collections import deque

def build_tree_by_levels(list_obj):
    if len(list_obj) == 0:
        return None

    tree = TreeNode(list_obj[0])
    q = deque()
    q.append(tree)
    i = 1
    while q:
        node = q.popleft()
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.left = TreeNode(list_obj[i])
                q.append(node.left)

        i+=1
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.right = TreeNode(list_obj[i])
                q.append(node.right)

        i+=1

    return tree
```

```python
tree = build_tree_by_levels(root)
```

```python
def isSymmetric_tree(tree):
    def isMirror(l, r):
        if not l and not r: # both thees are None
            return True

        if not l or not r: # one of them is None
            return False 

        return l.val == r.val and isMirror(l.right, r.left) and isMirror(l.left, r.right)

    return isMirror(tree.left, tree.right)

isSymmetric_tree(tree)
```
