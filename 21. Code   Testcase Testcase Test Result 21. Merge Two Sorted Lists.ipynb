21. Merge Two Sorted Lists

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        head = ListNode(0)
        node = head
        
        while l1 and l2:   
            if l2.val < l1.val:
                node.next = ListNode(l2.val)
                l2 = l2.next        

            else:
                node.next = ListNode(l1.val)
                l1 = l1.next
                
            node = node.next

        while l1:
            node.next = ListNode(l1.val)
            l1 = l1.next
            node = node.next

        while l2:
            node.next = ListNode(l2.val)
            l2 = l2.next
            node = node.next
            
        return head.next
            
```

```python
# Input: 
list1 = [1,2,4]
list2 = [1,3,4]
# Output: [1,1,2,3,4,4]
```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# Help functions
def build_linked_list(l):    
    node = ListNode(0)
    head = node

    for i in l:
        node.next = ListNode(i)
        node = node.next

    return head.next

def list_represent(l):
    res = []
    while l:
        res.append(l.val)
        l = l.next

    return res
    
```

```python
head1 = build_linked_list(list1)
head2 = build_linked_list(list2)
```

```python
print(list_represent(head1))
print(list_represent(head2))
```

```python
def merge_two_lists(l1, l2):
    head = ListNode(0)
    node = head
    
    while l1 and l2:   
        if l2.val < l1.val:
            node.next = ListNode(l2.val)
            l2 = l2.next        

        else:
            node.next = ListNode(l1.val)
            l1 = l1.next
            
        node = node.next

    while l1:
        node.next = ListNode(l1.val)
        l1 = l1.next
        node = node.next

    while l2:
        node.next = ListNode(l2.val)
        l2 = l2.next
        node = node.next
        
    return head.next
```

```python
merged = merge_two_lists(head1, head2)
print(list_represent(merged))
```
