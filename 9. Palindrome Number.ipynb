9. Palindrome Number

Given an integer x, return true if x is a 
palindrome
, and false otherwise.

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        res = 0
        reserve = x
        if x < 0:
            return False
        
        while x>0:
            last = x % 10
            x = x // 10
            res = res * 10 + last
        
        return res == reserve
```
