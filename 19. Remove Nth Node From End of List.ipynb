19. Remove Nth Node From End of List

Given the head of a linked list, remove the nth node from the end of the list and return its head.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """
        dummy = ListNode(0, head)
        fast = dummy
        slow = dummy
        i = 0
        while fast:            
            fast = fast.next
            if i >= n + 1:
                slow = slow.next

            i+=1

        if i > n:      
            slow.next = slow.next.next
        return dummy.next
```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# Input: 
head = [1] 
n = 1
# Output: []
```

```python
# Help function to build a linked list
from collections import deque

def build_linked_list(l):
    if len(l) == 0:
        return None
    q = deque()
    head = ListNode(l[0])
    q.append(head)
    i = 1
    while i < len(l):
        node = q.popleft()
        node.next = ListNode(l[i])
        q.append(node.next)
        i+=1

    return head

linked_head = build_linked_list(head)
```

```python
# Help function to print linked list

def print_linked_list(l_list):
    if not l_list:
        return []
        
    res = [l_list.val]
    node = l_list.next
    while node:        
        res.append(node.val)
        node = node.next

    return res

print_linked_list(linked_head)
```
