95. Unique Binary Search Trees II

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_all_possible_bst(elems_list):
    if len(elems_list) == 0:
        return [None]
    
    trees = []
   
    for root_idx in range(len(elems_list)):        
        left_elems = elems_list[:root_idx]
        right_elems = elems_list[root_idx + 1 :]

        left_trees = build_all_possible_bst(left_elems)
        right_trees = build_all_possible_bst(right_elems)

        for l in left_trees:
            for r in right_trees:
                tree = TreeNode(elems_list[root_idx], l, r)           
                trees.append(tree)
                
    return trees
```

```python
# Helper function for representation of Tree objects into the List of Tree values by levels
from collections import deque

def print_tree_by_levels(tree_node):
    if not tree_node:
        return []
    
    q = deque()
    q.append(tree_node)

    res = []
    while q:
        node = q.popleft()
        if node is None:
            res.append('null')
        else:
            res.append(node.val)
            
            
            q.append(node.left)
            q.append(node.right)


    while res[-1] == 'null':
        res.pop()
    return res          
```

```python
# LAUNCH

#input
n = 4
elements = list(range(1, n+1))

# build the List of unique BSTs
for tree in build_all_possible_bst(elements):
    print(print_tree_by_levels(tree))

```
