96. Unique Binary Search Trees

Given an integer n, return the number of structurally unique BST's (binary search trees) which has exactly n nodes of unique values from 1 to n.

```python
class Solution(object):
    def numTrees(self, n):
        results = [0] * (n+1)    
        results[0] = results[1] = 1 # base cases

        for n_nodes in range(2, n+1, 1):
            # print('n_nodes', n_nodes)
            for root_idx in range(n_nodes):
                left_tree = root_idx 
                right_tree = n_nodes - root_idx - 1
                # print(f'root_idx {root_idx}, left_tree {left_tree}, right_tree {right_tree}')
                results[n_nodes] += results[left_tree] * results[right_tree]

        return results[n]
        
```
