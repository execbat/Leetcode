108. Convert Sorted Array to Binary Search Tree

Given an integer array nums where the elements are sorted in ascending order, convert it to a 
height-balanced
 binary search tree.

```python
class Solution(object):
    def sortedArrayToBST(self, l):
        def helper(ptr_l, ptr_r):
            if ptr_l > ptr_r:
                return None
                
            ptr_m = (ptr_l + ptr_r)//2
            
            tree = TreeNode(l[ptr_m])
            tree.left = helper(ptr_l, ptr_m-1)
            tree.right = helper(ptr_m + 1, ptr_r)
            return tree      

        return helper(0, len(l) - 1)
```

```python
nums = [-10,-3,0,5,9]
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def build_bst(l):
    if len(l) == 0:
        return None

    mid_idx = len(l)//2
    tree = TreeNode(l[mid_idx])
    left_tree_elems = l[:mid_idx]
    right_tree_elems = l[mid_idx + 1 :]
    tree.left = build_bst(left_tree_elems)
    tree.right = build_bst(right_tree_elems)
    return tree

tree = build_bst(nums)
tree
```

```python
# solution faster

def build_bst(l):
    def helper(ptr_l, ptr_r):
        if ptr_l > ptr_r:
            return None
            
        ptr_m = (ptr_l + ptr_r)//2
        
        tree = TreeNode(l[ptr_m])
        tree.left = helper(ptr_l, ptr_m-1)
        tree.right = helper(ptr_m + 1, ptr_r)
        return tree      

    return helper(0, len(l) - 1)

tree = build_bst(nums)
tree
```

```python
# Helper function to make list representation of the Tree by levels

from collections import deque

def tree_traversal_by_levels(tree):
    if not tree:
        return None

    res_list = []
    q = deque()
    q.append(tree)
    res_list.append(tree.val)

    while q:
        node = q.popleft()
        if node.left:
            q.append(node.left)
            res_list.append(node.left.val)

        else:
            res_list.append("null")

        if node.right:
            q.append(node.right)
            res_list.append(node.right.val)

        else:
            res_list.append("null")

    while res_list[-1] == "null":
        res_list.pop()

    return res_list
tree_traversal_by_levels(tree)
# [0,-10,5,null,-3,null,9]
```
