104. Maximum Depth of Binary Tree

Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

```python
class Solution(object):
    def maxDepth(self, root):
        if not root:
            return 0

        return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1
```

```python
root = [3,9,20,"null","null",15,7]
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# Help function to build tree from list representation by levels
from collections import deque

def build_tree_by_levels(list_obj):
    if len(list_obj) == 0:
        return None

    tree = TreeNode(list_obj[0])
    q = deque()
    q.append(tree)
    i = 1
    while q:
        node = q.popleft()
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.left = TreeNode(list_obj[i])
                q.append(node.left)

        i+=1
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.right = TreeNode(list_obj[i])
                q.append(node.right)

        i+=1

    return tree
```

```python
tree = build_tree_by_levels(root)
```

```python
# Recursive solution

def getTreeDepth(tree):
    if not tree:
        return 0

    return max(getTreeDepth(tree.left), getTreeDepth(tree.right)) + 1

getTreeDepth(tree)
```

```python
# Queue solution

def getTreeDepth_queue(tree):
    if not tree:
        return 0
        
    q = deque()
    q.append(tree)
    depth = 0
    while q:
        nodes_in_level = len(q)        
        depth += 1
        for _ in range(nodes_in_level):
            node = q.popleft()
    
            if node.left:
                q.append(node.left)
    
            if node.right:
                q.append(node.right)

    return depth

getTreeDepth_queue(tree)          
```
