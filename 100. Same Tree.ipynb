100. Same Tree

Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

```python
class Solution(object):
    def isSameTree(self, p, q):
        if p is None and q is None: # both trees are None
            return True

        if p is None or q is None: # only one of two is None
            return False

        # here we both trees are not None
        # checking their values and recursively check left and right subtrees
        return p.val == q.val and self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
```

```python
p = [1,2,3]
q = [1,2,3]
```

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# Help function to build tree from list representation by levels
from collections import deque

def build_tree_by_levels(list_obj):
    if len(list_obj) == 0:
        return None

    tree = TreeNode(list_obj[0])
    q = deque()
    q.append(tree)
    i = 1
    while q:
        node = q.popleft()
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.left = TreeNode(list_obj[i])
                q.append(node.left)

        i+=1
        if i < len(list_obj):
            if list_obj[i] != "null":
                node.right = TreeNode(list_obj[i])
                q.append(node.right)

        i+=1

    return tree
```

```python
tree1 = build_tree_by_levels(p)
tree2 = build_tree_by_levels(q)
```

```python
def the_same_trees(tree1, tree2):    
    if tree1 is None and tree2 is None: # both trees are None
        return True

    if tree1 is None or tree2 is None: # only one of two is None
        return False

    # here we both trees are not None
    # checking their values and recursively check left and right subtrees
    return tree1.val == tree2.val and the_same_trees(tree1.left, tree2.left) and the_same_trees(tree1.right, tree2.right)

the_same_trees(tree1, tree2)
```
